## Leer audio usando scipy

```python
import scipy.io.wavfile as sc
```

```python
[F1, data1]=sc.read('data/forward/0a2b400e_nohash_1.wav')
plt.figure(2)
plt.plot(data1)
plt.title('Reading sound with scipy')
plt.ylabel('Amplitud')
plt.xlabel('Tiempo')
```
